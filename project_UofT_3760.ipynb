### 2: Remove custom dlp_tokens

```python
def remove_custom_dlp_tokens(text):
    # print("[INFO] RAW text ",text)
    # print("[INFO] list text ",text.split())
    text=str(text)
    clean=[word for word in text.split() if word[0]!="[" and word[-1]!="]"]
    output=" ".join(clean)
    # print("[INFO] output text ",output)
    return output
```

```python
final["clean_text"] = final["text"].apply(lambda x:remove_custom_dlp_tokens(x)) 
final["transcript"] = final["text"].apply(lambda x:remove_custom_dlp_tokens(x)) 
```

```python
final["clean_text"] = final["text"].apply(lambda x:re.sub("[\[].*?[\]]","",x)) 
final["transcript"] = final["text"].apply(lambda x:re.sub("[\[].*?[\]]","",x)) 
```

### Convert DF to JSON

```python


def df_to_json(df):

    """
    function to convert dataframe to json format for gtm

    Args:
        df: input dataframe for converting to json format

    return:
        None
    """

    # convert speaker keys and add user id column
    df["user_id"] = df["role"].map({"AGENT": 1, "END_USER": 2})

    # get all session ids
    all_sid = df["Count_id"].unique()

    with tqdm(total=len(all_sid)) as pbar:

        # itterate on all session id
        for sid in all_sid:

            # create subset df for each session id
            sub_df = df[df["Count_id"] == sid].reset_index(drop=True)

            session_json = dict()
            session_json["entries"] = []

            for idx, i in sub_df.iterrows():
                row = dict()
                row["start_timestamp_usec"] = i["Time"]
                row["text"] = i["text"]
                row["role"] = i["role"]
                row["user_id"] = i["user_id"]
                session_json["entries"].append(row)

            # write json to local path
            json_object = json.dumps(session_json, indent=4)
            with open("Data/NEW/INPUT_JSON/train_clean/{}.json".format(sid), "w") as outfile:
                outfile.write(json_object)
            pbar.update(1)

    print("[INFO] GCS dataframe to Json Done...")
```
